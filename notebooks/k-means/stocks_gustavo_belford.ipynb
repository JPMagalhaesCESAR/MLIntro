# Stocks k-means by Gustavo Belford

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline
data = pd.read_csv("sample_stocks.csv")

data.info()
data.describe()
```

```python
f1 = data['returns'].values
f2 = data['dividendyield'].values
X = np.array(list(zip(f1, f2)))
plt.scatter(f1, f2, c='black', s=7)
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=3)
kmeans = kmeans.fit(X)
labels = kmeans.predict(X)
centroids = kmeans.cluster_centers_

print(centroids)

f1 = data['returns'].values
f2 = data['dividendyield'].values
X = np.array(list(zip(f1, f2)))
plt.scatter(f1, f2, c='black', s=7)

for i in range(3):
    plt.scatter(centroids[i][0], centroids[i][1],c='blue', s=100)
```

```python
plt.scatter(X[:, 0], X[:,1], s = 100, c = kmeans.labels_)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 300, c = 'blue',label = 'Centroids')
plt.xlabel('Returns')
plt.ylabel('DividendYield')
 
plt.show()
```

```python
coto = []
for i in range(1, 10):
    kmeans = KMeans(n_clusters = i)
    kmeans.fit(X)
    print(i,kmeans.inertia_)
    coto.append(kmeans.inertia_)  
plt.plot(range(1, 10), coto)
plt.show()
```
