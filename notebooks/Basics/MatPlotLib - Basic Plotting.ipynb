# Basic Plotting using MatPlotLib

- Create an array from 0 to 1, 0.1 step

```python
import numpy as np
x = np.arange(0, 1, 0.1)
print x
```

- Create a linear function and plot it

```python
import matplotlib.pyplot as plt
%matplotlib inline

y = x*2 + 5
plt.plot(x, y)
```

- Create a x squared function and plot it

```python
y2 = x**2 + 5
plt.plot(x, y2)
```

- Create a senoid and plot it

```python
from numpy import pi, sin

x = np.arange(0, 2*pi, 0.01)
y = sin(x)
plt.plot(x, y)
```

- Create two senoids and plot it

```python
from numpy import pi, sin, cos

x = np.arange(0, 2*pi, 0.01)
y = sin(x)
ax = plt.plot(x, np.matrix([sin(x), cos(x)]).T)
```

- Create a random 'standard normal' distribution and plot its histogram
