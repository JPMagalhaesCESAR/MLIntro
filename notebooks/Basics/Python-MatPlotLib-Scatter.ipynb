# Python MatPlotLib Scatter Introduction

- Create 50 random examples (x, y) and display

```python
import numpy as np
import matplotlib.pyplot as plt

N = 50
x = np.random.rand(N)
y = np.random.rand(N)

plt.scatter(x, y)
```

- Set the figure size and the alpha of the points

```python
plt.figure(figsize=(10,10))
plt.scatter(x,y,alpha=0.5)
```

- Display the points using different collors

```python
plt.figure(figsize=(10,10))
colors = np.random.rand(N)
plt.scatter(x,y,alpha=0.5, c=colors)
```

- Use different sizes for each point

```python
plt.figure(figsize=(10,10))
area = (30 * np.random.rand(N))**2 # 0 to 15 point radii
plt.scatter(x,y,alpha=0.5, c=colors, s=area)
```

- Order the set in x and show it as two different cateories

```python
plt.figure(figsize=(10,10))

x.sort()

x1 = x[:25]
y1 = y[:25]

x2 = x[25:]
y2 = y[25:]

# plt.subplot(121)
plt.scatter(x1, y1, alpha=0.5, c=colors[:25], marker=">", s=area[:25])

# plt.subplot(122)
plt.scatter(x2, y2, alpha=0.5, c=colors[25:], marker="+", s=area[25:])

plt.show()
```

- Display each different subset as a sub image

```python
plt.figure(figsize=(10,10))

plt.subplot(121)
plt.scatter(x1, y1, alpha=0.5, c=colors[:25], marker=">", s=area[:25])

plt.subplot(122)
plt.scatter(x2, y2, alpha=0.5, c=colors[25:], marker="+", s=area[25:])

plt.show()
```
